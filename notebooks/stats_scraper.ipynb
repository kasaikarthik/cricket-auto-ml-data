```python
top_bowlers_master = {}

for season, url in bowling_urls.items():
    driver.get(url)
    print(season)
    
    html = driver.page_source.encode('utf-8')
    page_num = 0
    content = driver.page_source
    page_soup = soup(driver.page_source, 'lxml')

    table = page_soup.findAll('tr', {'class':'ds-bg-ui-fill-translucent'})
    labels = page_soup.findAll('span', {'class':'ds-cursor-pointer'})
    
    all_labels = []
    for i in range(14):
        all_labels.append(labels[i].text)
    
    top_bowlers = []
    count = 0

    for item in table:
        json_obj = {}
        table_objects = item.findAll('td')
        for count in range(len(table_objects)-1):
            json_obj[all_labels[count]] = table_objects[count].text
        json_obj['team'] = json_obj['Player'].split(' (')[1].replace(')', '')
        json_obj['bowler_surname'] = json_obj['Player'].split(' ')[-2]
        top_bowlers.append(json_obj)

    top_bowlers_pd = pd.DataFrame.from_dict(top_bowlers)
    top_bowlers_master[season] = top_bowlers_pd
```

```python
ipl_matches = pd.read_csv('ipl_all_seasons_v6.csv')
```

```python
ipl_matches['home_key_bowler'] = ipl_matches['home_key_bowler'].str.replace('*', '')
ipl_matches[['home_key_bowler_1', 'home_key_bowler_2']] = ipl_matches['home_key_bowler'].str.split(', ', expand=True)
```

```python
ipl_matches['home_key_batter'] = ipl_matches['home_key_batter'].str.replace('*', '')
ipl_matches[['home_key_batter_1', 'home_key_batter_2']] = ipl_matches['home_key_batter'].str.split(' , ', expand=True)
```

```python
ipl_matches['away_key_batter'] = ipl_matches['away_key_batter'].str.replace('*', '')
ipl_matches[['away_key_batter_1', 'away_key_batter_2']] = ipl_matches['away_key_batter'].str.split(' , ', expand=True)
```

```python
ipl_matches['away_key_bowler'] = ipl_matches['away_key_bowler'].str.replace('*', '')
ipl_matches[['away_key_bowler_1', 'away_key_bowler_2']] = ipl_matches['away_key_bowler'].str.split(', ', expand=True)
```

```python
home_key_batter_2 = ipl_matches.groupby(['home_key_batter_2', 'season']).count().reset_index()[['home_key_batter_2','season','id']].rename(columns={'id':'home_key_batter_2_count', 'home_key_batter_2':'batter'})
home_key_batter_1 = ipl_matches.groupby(['home_key_batter_1', 'season']).count().reset_index()[['home_key_batter_1','season','id']].rename(columns={'id':'home_key_batter_1_count', 'home_key_batter_1':'batter'})

away_key_batter_2 = ipl_matches.groupby(['away_key_batter_2', 'season']).count().reset_index()[['away_key_batter_2','season','id']].rename(columns={'id':'away_key_batter_2_count', 'away_key_batter_2':'batter'})
away_key_batter_1 = ipl_matches.groupby(['away_key_batter_1', 'season']).count().reset_index()[['away_key_batter_1','season','id']].rename(columns={'id':'away_key_batter_1_count', 'away_key_batter_1':'batter'})
```

```python
home_key_bowler_2 = ipl_matches.groupby(['home_key_bowler_2', 'season']).count().reset_index()[['home_key_bowler_2','season','id']].rename(columns={'id':'home_key_bowler_2_count', 'home_key_bowler_2':'bowler'})
home_key_bowler_1 = ipl_matches.groupby(['home_key_bowler_1', 'season']).count().reset_index()[['home_key_bowler_1','season','id']].rename(columns={'id':'home_key_bowler_1_count', 'home_key_bowler_1':'bowler'})

away_key_bowler_2 = ipl_matches.groupby(['away_key_bowler_2', 'season']).count().reset_index()[['away_key_bowler_2','season','id']].rename(columns={'id':'away_key_bowler_2_count', 'away_key_bowler_2':'bowler'})
away_key_bowler_1 = ipl_matches.groupby(['away_key_bowler_1', 'season']).count().reset_index()[['away_key_bowler_1','season','id']].rename(columns={'id':'away_key_bowler_1_count', 'away_key_bowler_1':'bowler'})
```

```python
home_batter_record = home_key_batter_1.merge(home_key_batter_2, how='outer', left_on=['batter', 'season'], right_on=['batter', 'season'])
away_batter_record = away_key_batter_1.merge(away_key_batter_2, how='outer', left_on=['batter', 'season'], right_on=['batter', 'season'])
```

```python
home_bowler_record = home_key_bowler_1.merge(home_key_bowler_2, how='outer', left_on=['bowler', 'season'], right_on=['bowler', 'season'])
away_bowler_record = away_key_bowler_1.merge(away_key_bowler_2, how='outer', left_on=['bowler', 'season'], right_on=['bowler', 'season'])
```

```python
all_batter_record = home_batter_record.merge(away_batter_record, left_on=['batter', 'season'], right_on=['batter', 'season'], how='outer')
```

```python
all_bowler_record = home_bowler_record.merge(away_bowler_record, left_on=['bowler', 'season'], right_on=['bowler', 'season'], how='outer')
```

```python
all_batter_record = all_batter_record.fillna(0)
all_batter_record['overall_key_count'] = all_batter_record['home_key_batter_1_count'] + all_batter_record['home_key_batter_2_count'] + all_batter_record['away_key_batter_1_count'] + all_batter_record['away_key_batter_2_count']
all_batter_record['home_key_count'] = all_batter_record['home_key_batter_1_count'] + all_batter_record['home_key_batter_2_count']
all_batter_record['away_key_count'] = all_batter_record['away_key_batter_1_count'] + all_batter_record['away_key_batter_2_count']
```

```python
all_bowler_record = all_bowler_record.fillna(0)
all_bowler_record['overall_key_count'] = all_bowler_record['home_key_bowler_1_count'] + all_bowler_record['home_key_bowler_2_count'] + all_bowler_record['away_key_bowler_1_count'] + all_bowler_record['away_key_bowler_2_count']
all_bowler_record['home_key_count'] = all_bowler_record['home_key_bowler_1_count'] + all_bowler_record['home_key_bowler_2_count']
all_bowler_record['away_key_count'] = all_bowler_record['away_key_bowler_1_count'] + all_bowler_record['away_key_bowler_2_count']
```

```python
all_batter_record = all_batter_record.sort_values(by=['overall_key_count', 'away_key_count'], ascending=False)
```

```python
all_batter_record.drop(['home_key_batter_1_count', 'home_key_batter_2_count', 'away_key_batter_1_count', 'away_key_batter_2_count'], axis=1, inplace=True)
```
